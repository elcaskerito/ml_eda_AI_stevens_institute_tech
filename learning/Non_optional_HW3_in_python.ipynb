```python
import numpy as np
import pandas as pd
from sklearn import neighbors
from sklearn.metrics import accuracy_score
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# import of our data 
trainingSet = pd.read_csv('HW_03_knn_BC_pfizer.csv', header=0)
```

```python
# read or our data 

trainingSet.head()

# divided our data into testing data (ie. 6th row) and training data (the remaining data)
testSet = trainingSet.iloc[[5, 11, 17,23, 29]]

df = trainingSet.drop(index=[5, 11, 17,23, 29])
```

```python
# drop all na
# df.dropna(inplace=True)
df.drop(columns='Row_N0', inplace=True)
```

```python
df.head()
```

```python
# dropping the sample attribute because it's not relative to our model
df.drop(columns='Sample', inplace=True)
```

```python
df.head()
```

```python
# define out feature (x) and lable (Y)

x=df.drop(columns='Class').copy()
y=df['Class']
```

```python
# create the test and train datasets for feature and labels

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20, random_state=6)
```

```python

# create an objet of the methode

knn = neighbors.KNeighborsClassifier()

# fit the model

knn.fit(x_train,y_train)
```

```python
# get the accuracy of the model on the test dataset
knn.score(x_test,y_test)
```

```python
k_values = [1,2,3]
# k_values = range(1,15)
 
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train, y_train)
    target_pred = knn.predict(x_test)
    accuracy = accuracy_score(y_test,target_pred ) 
    print(f'Accuracy of model with k = {k}: {accuracy}')
    print('')
```
