```python

file_path='/home/elcaskerito/Documents/Stevens Jupiter/new/ml_eda_AI_stevens_institute_tech/learning/hepatitis_C2.csv'

# # Read CSV file with appropriate handling of missing values
df = pd.read_csv(file_path, na_values='?')


print(df.describe())

df.head(30)
```

```python


attr = df.drop(columns = ['Row']) 
target = attr['Class']

trainSet = attr.drop(columns = ['Class']) 



print(trainSet)



print('\n--------------------------------------- *test set start here*-------------------------- \n')



print(target.head(5))
# df.head(10)
# print(testSet)
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize MinMaxScaler
scaler = MinMaxScaler()
# Fit and transform the data
trainSet = pd.DataFrame(scaler.fit_transform(trainSet), columns=trainSet.columns)

print('\n--------------------------------------- *Scalling the data*-------------------------- \n')
trainSet.head()


```

```python

train_trainSet,test_trainSet, train_testSet, test_testSet = train_test_split(trainSet, target,test_size = 0.16, random_state = 13, shuffle = True)

```

```python
# from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
k_values = [1,2,3]


for k in k_values:
    knn = KNeighborsClassifier(n_neighbors = k)
#     knn = KNeighborsClassifier(n_neighbors = k,weights='distance')
    knn.fit(train_trainSet, train_testSet)
    target_pred = knn.predict(test_trainSet)
#     accuracy = round(np.mean(target_test==target_pred ) * 100, 2)
    accuracy = accuracy_score(test_testSet,target_pred ) 
    print(f'Accuracy of model with k = {k}: {accuracy}')
    print('')
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(test_testSet, target_pred)

print('Confusion Matrix')
print(confusion_matrix(test_testSet, target_pred))
print()
print('Accuracy score')
print(accuracy_score(test_testSet, target_pred))
print()
print('Classification Report')
print(classification_report(test_testSet, target_pred))
```
